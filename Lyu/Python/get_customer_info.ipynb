```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel("../Materials/input_node.xlsx")
```

```python
cs = data[(data.ID >= 1) & (data.ID <= 1000)]
```

```python
# convert 00:00:00 to minutes away from 8:00 AM.
for i in range(1,1001):
    cs.at[i,"first_receive_tm"] = (cs.loc[i,"first_receive_tm"].hour-8)*60 + cs.loc[i,"first_receive_tm"].minute
    cs.at[i,"last_receive_tm"] = (cs.loc[i,"last_receive_tm"].hour-8)*60 + cs.loc[i,"last_receive_tm"].minute
```
