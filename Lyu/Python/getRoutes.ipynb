```python
# get the distance from i to j
def getDist (fromNode,toNode):
    dist = dist_and_time[(dist_and_time.from_node == fromNode) & (dist_and_time.to_node == toNode)]
    return int(dist.distance)

# get the traval time from i to j
def getTime (fromNode,toNode):
    dist = dist_and_time[(dist_and_time.from_node == fromNode) & (dist_and_time.to_node == toNode)]
    return dist.spend_tm

# get weight of node i
def getWeit (i):
    # check if i is allowed
    if i in range(1,51):
        row = data[data.ID == 1]
        return row.pack_total_weight
    # if i is not allowed
    else: 
        print('input is illegal')
        
# get volume of node i
def getVolu (i):
    # check if i is allowed
    if i in range(1,51):
        row = data[data.ID == 1]
        return row.pack_total_volume
    # if i is not allowed
    else: 
        print('input is illegal')
```

# Get the routes

```python
# Initialization
routes=[0]
custID = list(range(1,51))
starNode = 0
toNode = 1
```

```python
for i in range(1,51):
    Dist = getDist(starNode,custID[0])
    print('\n\ntemp',i,': from',starNode,'to ?')
    
    for j in custID:
        if (j != starNode):
            if getDist(starNode,j) <= Dist:
                toNode = j 
                Dist = getDist(starNode,j)
    print('the nearest node of',starNode,'is',toNode,',the distance is',Dist)
    
    custID.remove(toNode)
    routes.append(toNode)
    print('routes are',routes)
    starNode = toNode

print("\n\nCaculation completed")
routes.append(0)
print('The final routes are:\n',routes)
```

# Cut the routes

```python
rout = routes
rout.remove(0)
```
