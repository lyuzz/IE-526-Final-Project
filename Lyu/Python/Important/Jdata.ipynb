```python
import pandas as pd
import random
import numpy as np
from datetime import datetime
data = pd.read_excel('../../Materials/input_node.xlsx')
dist_and_time = pd.read_csv('../../Materials/input_distance-time.txt')
```

```python
# The capacity of truck
max_weight = 2.5
max_volume = 16
max_drive = 120000
max_time = 720
cost_distance = 14
cost_vehicle = 300
cost_wait = 24
```

```python
# get the distance from i to j
def distance (fromNode,toNode):
    if fromNode == toNode:
        print("Illegal (distance): i = j.")
    else:
        dist = dist_and_time[(dist_and_time.from_node == fromNode) & (dist_and_time.to_node == toNode)]
        return int(dist.distance)

# get the traval time from i to j
def time (fromNode,toNode):
    if fromNode == toNode:
        print("Illegal (time): i = j.")
    else:
        dist = dist_and_time[(dist_and_time.from_node == fromNode) & (dist_and_time.to_node == toNode)]
        return int(dist.spend_tm)

# get weight of node i
def weight (i):
    if i in range(1,1001):             # check if i is allowed
        row = data[data.ID == i]
        return float(row.pack_total_weight)
    else:                            # if i is not allowed     
        print("Illegal (weight): Customer ID should be between 1 to 1000.")
        
# get volume of node i
def volume (i):
    if i in range(1,1001):
        row = data[data.ID == i]
        return float(row.pack_total_volume)
    else: 
        print("Illegal (volume): Customer ID should be between 1 to 1000.")
```

```python
def getInd(customerID):
    # Initialization
    custID = customerID[:] # make a copy of customerID
    nbCustomer = len (customerID)
    route = []
    starNode = 0
    toNode = 1

    # get a route by the least distance
    for attemp in range(1,nbCustomer + 1):
        Dist = distance(starNode,custID[0])

        for node in custID:
            if (node != starNode):
                if distance(starNode,node) <= Dist:
                    toNode = node 
                    Dist = distance(starNode,node)

        custID.remove(toNode)
        route.append(toNode)
        starNode = toNode
        print("Attemp:",attemp,"/",nbCustomer)
        
    # print('The final routes are:\n',route,"\n")
    return route
```

```python
def ind2routes(ind):
    route = []
    routes = []
    fromNode = 0
    weig = 0
    volu = 0
    Time = 0

    for nextNode in ind:
        # If next customer could fit.
        if ((weig+weight(nextNode) <= max_weight)&
            (volu+volume(nextNode) <= max_volume)&
            (Time+time(fromNode,nextNode)+time(nextNode,0) <= max_time)):
            
            route.append(nextNode)
            weig = weig + weight(nextNode)
            volu = volu + volume(nextNode)
            Time = Time + time(fromNode,nextNode)
            fromNode = nextNode
            
        # If next customer doesn't fit, let truck go back.
        else:
            routes.append(route)
            # Reset all of the capacity and make assignment for next truck.
            route = []
            weig = 0
            volu = 0
            Time = 0
            fromNode = 0
            # If this node could fit in a new truck?
            if ((weight(nextNode) <= max_weight)&
                (volume(nextNode) <= max_volume)&
                (time(fromNode,nextNode)+time(nextNode,0) <= max_time)): 
                
                Time = Time + time(fromNode,nextNode)
                weig = weig + weight(nextNode)
                volu = volu + volume(nextNode)
                route.append(nextNode)
                fromNode = nextNode

            else:
                print("Error! There is no feasible solution because node",nextNode,"can't fit in a new truck")
                break
                
    routes.append(route) 
    return routes
```

```python
def nearestStation(customerID):
    station_index = 1001
    for station in range (1001,1101):
        if distance(customerID,station) <= distance(customerID,station_index):
            station_index = station
    return station_index
```

```python
def repairRoutes(routes):
    newRoutes = []
    i = 1
    for route in routes:
        distRemain = max_drive
        newRoute = []
        fromNode = 0
        lastNode = 0
        for nextNode in route:
            # vehicle can drive to the next customer
            if distRemain >= distance(fromNode,nextNode):
                newRoute.append(nextNode)
                distRemain = distRemain - distance(fromNode,nextNode)
                lastNode = fromNode
                fromNode = nextNode
            else:
                # find the least charging station
                station_index = nearestStation(fromNode)
                # if vehilce are able to go to the nearest charge station
                if distRemain >= distance(fromNode,station_index):
                    # go to the charge station
                    newRoute.append(station_index)
                    newRoute.append(nextNode)
                    distRemain = max_drive - distance(station_index,nextNode)
                    lastNode = station_index
                    fromNode = nextNode
                else:
                    station_index = nearestStation(lastNode)
                    if distRemain + distance(lastNode,fromNode) >= distance(lastNode,station_index):
                        newRoute.insert(len(newRoute)-1,station_index)
                        newRoute.append(nextNode)
                        distRemain = max_drive - distance(station_index,fromNode) - distance(fromNode,nextNode)
                        lastNode = fromNode
                        fromNode = nextNode
                    else:
                        print(i,"fuck!",route)
                        print(distRemain + distance(lastNode,fromNode),distance(lastNode,station_index))

        # add into new routes
        newRoutes.append(newRoute)
        i += 1
    return newRoutes
```

```python
def evalateRoutes(routes):
    i = 1
    cost = 0
    flag = 0
    for route in routes:
        weig = 0
        volu = 0
        dist = 0
        Time = 0
        fromNode = 0
        
        for node in route:
            weig = weig + weight(node)
            volu = volu + volume(node)
            dist = dist + distance(fromNode,node)
            Time = Time + time(fromNode,node)
            fromNode = node

        dist = dist + distance(fromNode,0)
        Time = Time + time(fromNode,0)

        if weig > max_weight or volu > max_volume or Time > max_time or dist > max_drive:
            print(i,route,"is not feasible!")
            print("weight,volume,time,distance:",weig,volu,Time,dist)
            flag = flag + 1
            i += 1
        else:
            print(i,route,"passed.")
            print("weight,volume,time,distance:",weig,volu,Time,dist)
            cost = cost + (dist * 14 / 1000) + 300
            i += 1
    if flag == 0:
        return cost
```
