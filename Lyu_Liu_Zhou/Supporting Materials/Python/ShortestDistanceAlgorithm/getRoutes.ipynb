# Input Data

```python
from datetime import datetime
import pandas as pd
data = pd.read_excel('../../RowData//input_node.xlsx')
dist_and_time = pd.read_csv('../../RowData//input_distance-time.txt')
```

```python
# get the distance from i to j
def getDist (fromNode,toNode):
    dist = dist_and_time[(dist_and_time.from_node == fromNode) & (dist_and_time.to_node == toNode)]
    return int(dist.distance)

# get the traval time from i to j
def getTime (fromNode,toNode):
    dist = dist_and_time[(dist_and_time.from_node == fromNode) & (dist_and_time.to_node == toNode)]
    return int(dist.spend_tm)

# get weight of node i
def getWeit (i):
    if i in range(1,1001):             # check if i is allowed
        row = data[data.ID == i]
        return float(row.pack_total_weight)
    else:                            # if i is not allowed     
        print('input of weight is illegal')
        
# get volume of node i
def getVolu (i):
    if i in range(1,1001):
        row = data[data.ID == i]
        return float(row.pack_total_volume)
    else: 
        print('input of volume is illegal')
```

# Get the best order of routes

```python
# Pick up customers in the order
# Set parameters
# nbCustomer = 50
# custID = list(range(1,nbCustomer + 1))
```

```python
# Given the customer ID
nbCustomer = 50
custID= [185,888,423,360,660,662,137,322,631,689,559,550,582,606,943,901,121,763,972,303,247
       ,17,140,101,255,851,645,977,601,810,406,400,182,484,276,359,947,215,616,221,299,833
       ,184,604,443,283,51,109,331,842]
```

```python
# Initialization
timeStarted = datetime.now()
routes = [0]
starNode = 0
toNode = 1

for i in range(1,nbCustomer + 1):
    Dist = getDist(starNode,custID[0])
    print('\nattemp',i,': from',starNode,'to ?')
    
    for j in custID:
        if (j != starNode):
            if getDist(starNode,j) <= Dist:
                toNode = j 
                Dist = getDist(starNode,j)
    print('the nearest node of',starNode,'is',toNode,',the distance is',Dist)
    
    custID.remove(toNode)
    routes.append(toNode)
    print('routes are',routes)
    starNode = toNode

print("\n\nCaculation completed")
routes.append(0)
print('The final routes are:\n',routes)
```

# Cut the routes

```python
# Remove distribution center O to get a set of customers.
if 0 in routes:  
    routes.remove(0)
    routes.remove(0)
    print("Removal completed!")
else:
    print("There is no distribution in this set")
```

```python
# The capacity of truck
max_weit = 2.5
max_volu = 16
max_driv = 200000
max_time = 720
```

```python
# Initialization
fromNode = 0
weight = 0
volume = 0
distance = 0
time = 0
routePlan = [0]
truck = 1
travelCost = 0
timeExceed = 0

for nextNode in routes:
    # If next customer could fit.
    if ((weight+getWeit(nextNode)<=max_weit)&(volume+getVolu(nextNode)<=max_volu)&(distance+getDist(fromNode,nextNode)+getDist(nextNode,0)<=max_driv)&(time<=max_time+getTime(fromNode,nextNode)+getTime(nextNode,0))):
        routePlan.append(nextNode)
        distance = distance + getDist(fromNode,nextNode)
        time = time + getTime(fromNode,nextNode)
        weight = weight + getWeit(nextNode)
        volume = volume + getVolu(nextNode)
        fromNode = nextNode  
    
    # If next customer doesn't fit, let truck go back.
    else:
        routePlan.append(0)
        time = time + getTime(fromNode,0)
        distance = distance + getDist(fromNode,0)
        travelCost = travelCost + distance * 14 /1000
        # Print the route of this truck.
        print("The route of truck",truck,"is",routePlan)
        print('Total weight is',weight)
        print('Total volume is',volume)
        print('Total traval time is',time)
        print('Total traval distance is',distance)
        print('\n')
        
        # Reset all of the capacity and make assignment for next truck.
        truck = truck + 1
        weight = 0
        volume = 0
        distance = 0
        time = 0
        routePlan = [0]
        fromNode = 0
        
        # If this "nextNode" could fit in a new truck?
        if ((weight<=max_weit)&(volume<=max_volu)&(distance+getDist(fromNode,nextNode)+getDist(nextNode,0)<=max_driv)&(time<=max_time+getTime(fromNode,nextNode)+getTime(nextNode,0))): 
            distance = distance + getDist(fromNode,nextNode)
            time = time + getTime(fromNode,nextNode)
            weight = weight + getWeit(nextNode)
            volume = volume + getVolu(nextNode)
            routePlan.append(nextNode)
            fromNode = nextNode
            
        else:
            print("Oh no!")
            break

# The last loop is completed and let this truck go home.
routePlan.append(0)
time = time + getTime(fromNode,0)
distance = distance + getDist(fromNode,0)

# caculate the cost
travelCost = travelCost + distance * 14 /1000
fixCost = truck * 300


timeFinished = datetime.now()
runTime = timeFinished - timeStarted
totalCost = fixCost + travelCost + timeExceed

# print the result
print("The route of truck",truck,"is",routePlan)
print('Total weight is',weight)
print('Total volume is',volume)
print('Total traval time is',time)
print('Total traval distance is',distance)
print('\n')
print("Completed!")
print("Run time =",runTime)
print("Total cost = ",totalCost)
    
```
